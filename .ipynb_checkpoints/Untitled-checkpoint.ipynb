```python
from pyspark.sql import SparkSession

# Inisialisasi sesi Spark
spark = SparkSession.builder.appName("FilterData").getOrCreate()

# Baca file CSV ke dalam DataFrame PySpark
data = spark.read.csv("LoanApprovalPrediction.csv", header=True, inferSchema=True)

# Filter baris dengan nilai "Indonesia" pada kolom tertentu
filtered_data = data.filter(data['nama_kolom'] == 'Indonesia')
filtered_data.show()

# Jangan lupa untuk menghentikan sesi Spark ketika selesai
spark.stop()
```

```python
import seaborn as sns
object_cols = [col_name for col_name, col_type in data.dtypes if col_type == "string"]

# Set up the Matplotlib figure
fig = plt.figure(figsize=(18, 36))
index = 1

# Iterate over object columns and create bar plots
for col in object_cols:
    col_counts = data.groupBy(col).count().orderBy("count", ascending=False).toPandas()
    
    plt.subplot(11, 4, index)
    plt.xticks(rotation=90)
    sns.barplot(x=col_counts[col], y=col_counts["count"])
    plt.title(col)
    
    index += 1

# Show the plots
plt.show()
```

```python
# Initialize a list to store StringIndexer models
indexer_models = []

# Iterate over object columns and perform label encoding
for col in object_cols:
    indexer = StringIndexer(inputCol=col, outputCol=col + "_index")
    model = indexer.fit(data)
    data = model.transform(data)
    indexer_models.append(model)

# Show the resulting DataFrame
data.show()

# To check the label encoding mapping
for model in indexer_models:
    print(f"Label encoding mapping for {model.getInputCol()}:")
    label_mapping = model.labels
    for i, label in enumerate(label_mapping):
        print(f"{label} -> {i}")
```

```python
from pyspark.sql.functions import mean, when, col

# Iterate over the columns and fill missing values with column mean
for col_name in data.columns:
    mean_value = data.select(mean(col(col_name))).collect()[0][0]
    data = data.withColumn(col_name, when(col(col_name).isNull(), mean_value).otherwise(col(col_name)))

# Check for missing values after filling
missing_counts = data.select([col(c).alias(c) for c in data.columns]).na.fill(0).agg(*[
    (count("*") - count(c)).alias(c) for c in data.columns
])

missing_counts.show()


```
